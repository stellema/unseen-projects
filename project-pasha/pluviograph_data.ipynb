```python
import pandas as pd
import numpy as np
```

```python
headings = pd.read_csv('pluviograph_data.csv', header=None, nrows=1)
headings = headings.dropna(axis=1).iloc[0].values
```

```python
headings
```

```python
df = pd.read_csv('pluviograph_data.csv', header=0)
df = df.iloc[:, 0:-2]
df.columns = headings
```

```python
def extract_location(df, location):
    """Extract columns corresponding to a location."""

    df_subset = df[location]
    new_header = df_subset.iloc[0]
    df_subset = df_subset[1:]
    df_subset.columns = new_header
    df_subset.set_index('Time', inplace=True)
    df_subset.columns = ['Cumulative Daily Rainfall',]
    df_subset['Cumulative Daily Rainfall'] = df_subset['Cumulative Daily Rainfall'].astype('float')

    return df_subset
```

```python
df_location = extract_location(df, 'Adamstown')
```

```python
df_location
```

```python
rainfall = df_location['Cumulative Daily Rainfall'].values[1:] - df_location['Cumulative Daily Rainfall'].values[0:-1]
rainfall = np.clip(rainfall, a_min=0, a_max=None)
rainfall = np.insert(rainfall, 0, df_location['Cumulative Daily Rainfall'].values[0])
df_location['Rainfall'] = rainfall
```
